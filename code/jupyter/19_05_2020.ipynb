## Control de Flujo

### Condicionales

```
a = 10
b = 30
if a < b or b < 20:
  print( "Primera respuesta" )
else:
  print( "Segunda respuesta" )
```

```
a = 10
b = 30
if a < b and b < 20:
  print( "Primera respuesta" )
elif a < b or b < 20:
  print( "Segunda respuesta" )
else:
  print( "Tercera respuesta" )
```

### Bucles condicionados

```
a,b = 0,1 # Inicialización de la sucesión de Fibonacci
print(a,end=' ')

while b<20:
  print(b, end=' ')
  a,b = b,a+b
```

```
acumulador = 0 # Inicialización de la sucesión de Fibonacci
indice = 1

while indice <= 10:
  acumulador = acumulador + 10 
  indice = indice + 1

print( acumulador )
```

## *Ejercicio 1*

## Ejercicio 2


**Algoritmo**

![Algoritmo](http://guillermopizarro.info/wp-content/uploads/sites/774/2020/06/12.png)

**Prueba de Escritorio**
![Prueba de Escritorio](http://guillermopizarro.info/wp-content/uploads/sites/774/2020/06/12b.png)

```
centinela = True

while centinela:
  cantidad_ = input("Ingresar la cantidad de números a sumar: ")
  cantidad = int(cantidad_)

  if cantidad > 1:
    centinela = False

i = 1
acumulador = 0

while i <= cantidad:
  numero_ = input( "Ingresar número a sumar: " )
  numero = int( numero_ )
  acumulador = acumulador + numero
  i = i + 1

print( acumulador )
```
