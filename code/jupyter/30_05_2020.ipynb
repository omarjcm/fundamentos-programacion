## Control de Flujo

### Bucles

```
a = [2, 3, 4, 5, 6, 7, 8, 100, 'a', 1000]

for i in range(len( a )):
  print( a[i] )
```

### Interrupción y Continuación

```
for k in range(2, 10):
  for i in range(2, k):
    if not k % i:
      print( k, ' es igual a ', i, '*', k//i )
      break
  else:
    print( k, ' es primo.' )
```

```
for i in range(1, 10):
  print(i)
else:
  print("Se ejecuta siempre")
```

```
for j in range(1, 5):
  for i in range(1, 10):
    print( i )
  else:
    print("Se ejecuta siempre")
```

```
k = 1; mynumber = 7
while( k < 10 ):
  if k == mynumber:
    break
  print( k, end=' ' )
  k += 1
else:
  print( "No" )
```

```
for k in range(10):
  if not k%2:
    print(k, ' es par.')
  else:
    pass
```

## Funciones

```
import math as m
def sumar(a, b):
  print( "La suma es: " + str( a + b ) )

sumar( 10, 20 )

def velocidad(espacio, tiempo):
  return (espacio / tiempo);

temp = velocidad(100, 1);
print( temp, 'Km/h ' )


def fibonacci(k):
  a, b = 0, 1
  print(a, end=' ')
  while(b < k):
    print(b, end=' ')
    a, b = b, a + b

fibonacci( 100000 )

```
