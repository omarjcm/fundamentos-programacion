## Variables Numéricas

En Python existe el tipado dinámico.

```
a = 1
b = 5.6

print(a)
print(b)
```

```
c = 6
d = 5
print( c//d )
```

```
e = 6.
f = 5.
print( e//f )
```

```
g = 6.
h = 5.
print( g/h )
```

```
j = 1
j = j + 1
print( j )

i = 1
i += 1
print( i )
```
