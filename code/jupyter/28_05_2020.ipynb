Intercambio sin tuplas, se necesita una variable extra (**temp**).

```
a_ = 0
b_ = 1

print( "a: " + str(a_) )
print( "b: " + str(b_) )

temp = b_
b_ = a_
a_ = temp

print( "\na: " + str(a_) )
print( "b: " + str(b_) )
```

Intercambio con tuplas, es directo.

```
a, b = 0, 1

print( "a: " + str(a) )
print( "b: " + str(b) )

# Intercambio
b, a = a, b

print( "\na: " + str(a) )
print( "b: " + str(b) )
```

## Booleanos y Comparaciones

Solo se consideran los elementos que son numéricos para las comparaciones.

```
[1, 'a'] < [0, 'b']
```

```
[0, 'a'] < [1, 'b']
```

```
[0, 'b'] < [1, 'a']
```

```
[0, 2, 4] <= [0, 7, 8]
```

```
[0, 2, 4] >= [0, 7, 8]
```

```
x = list('Abadacadabra')
print( x )

x = sorted( x )

print( x )
```

```
x = list('AbaDaCadaBra')

# Presenta toda la lista
print( x )
z = sorted(x)

# Presenta la lista ordenada de manera ascendente (primero las mayúsculas, luego las minúsculas)
print( z )
y = sorted(x, reverse=True)

# Presenta la lista ordenada de manera descendente (primero las minúsculas, luego las mayúsculas)
print( y )

# Presenta la lista ordenada de manera descendente (sin importar si es mayúscula o minúscula)
x = sorted(x, key=str.lower, reverse=True)

print( x )
```

## Control Flujo

### Bucles

```
for i in range(3):
  print( i )
```

```
a = list(range(10))
print( a )
a = list(range(5, 10))
print( a )
```

```
for i in range(3):
  for k in range(2):
    print( "i: " + str(i) + " k: " + str(k) + ' --- Suma: ' + str(i+k) )
```

```
a = 'hola mundo ups'

# split es una función para separar palabras
for b in a.split():
  for s in b:
    print(s, end=' ')
  print()

# se presentan las letras, sin tener en cuenta las palabras de la frase
for b in a:
  print(b, end=' ')
```
