## Slicing

Accediendo a elementos.

```
a = [9, 8, 7, 6, 5, 4, 3, 2, 1, 0]

# desde el índice 2 hasta el índice 4
print( a[2:5] )
# desde el índice 0 hasta el índice 5
print( a[:6] )
# desde el índice 6 hasta el final
print( a[6:] )
# desde el índice 1 hasta el índice 6, de dos en dos.
print( a[1:7:2] )
print( a[20:30] )
# se contabiliza desde el final hacia el inicio (el final es siempre índice -1)
print( a[-5:-3] )
# En positivo sería desde el índice 3 hasta el índice 6.
# En negativo sería desde el índice -7 hasta el índice -4
print( a[3:-3] )
```

Eliminando elementos.

```
a = [9, 8, 7, 6, 5, 4, 3, 2, 1, 0]
a[1:3] = []
print( a )
```

Modificando elementos.

```
a = [9, 8, 7, 6, 5, 4, 3, 2, 1, 0]
a[2:5] = [-1, -2, -3, -4]

print( a )
```

```
a = [9, 8, 7, 6, 5, 4, 3, 2, 1, 0]
a[1:1] = [0, 10, 20, 30]

print( a )
```

```
a = [9, 8, 7, 6, 5, 4, 3, 2, 1, 0]
a[1:2] = [0, 10, 20, 30]

print( a )
```
