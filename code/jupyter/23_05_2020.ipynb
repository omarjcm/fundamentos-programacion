## Ejercicios

### Ejercicio 1

Desarrollar un programa que determine el sueldo para cada uno de los empleados de una empresa. La empresa paga “tiempo normal”, para las primeras 40 horas trabajadas por el empleado y paga horas extra de 1.5 veces el valor de hora, para todas las horas exceso, trabajadas entre las 41 y 60 horas.  A partir de las 61 horas la empresa paga a 2.5 veces el valor de hora, y adicionalmente un bono equivalente al 10% del sueldo total.

**Planteamiento del Problema**

![Solución Alternativa](http://guillermopizarro.info/wp-content/uploads/sites/774/2020/05/ejercicio_03_planteamiento-2.png)

```
# Entrada
horas_ = input( "Ingresar las horas de trabajo: " )
horas = int( horas_ )

# Procesamiento
valorxhora = 10
sueldo_total = 0.0

if (horas > 0 and horas <= 40):
  sueldo_total = horas * valorxhora
elif (horas > 40 and horas <= 60):
  horas_extras = horas - 40
  sueldo_total = 40 * valorxhora + horas_extras * valorxhora * 1.5
elif (horas > 60):
  horas_extras = horas - 60
  sueldo = (40 * valorxhora) + (20 * valorxhora * 1.5) + (horas_extras * valorxhora * 2.5)
  bono = sueldo * 0.10
  sueldo_total = sueldo + bono

# Salida
print( "Sueldo Total: $ " + str(sueldo_total) )
```
