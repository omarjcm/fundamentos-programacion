### Ejercicio 1

Una empresa química paga a su personal de ventas en base a comisiones. El personal de ventas recibe 150 dólares por semana más un 5% de comisión mensual si la suma de ventas está entre 1 y 1000 dólares, 9 % de comisión mensual si las ventas están entre 1001 y 5000 dólares, si las ventas superan los 5000 dólares ganara un porcentaje del 12 %, adicionalmente un bono del 10 % del sueldo mensual que no incluye comisión. Desarrolle un programa que introduzca la venta neta mensual del vendedor, y calcule su salario mensual.
Nota: El porcentaje de la comisión se saca con respecto al total de las ventas mensuales.

**Planteamiento del Problema**

![Ventas](http://guillermopizarro.info/wp-content/uploads/sites/774/2020/06/24.png)

**Algoritmo**

![Algoritmo](http://guillermopizarro.info/wp-content/uploads/sites/774/2020/06/23.png)

```
# Entrada
venta_ = input( "Ingresar el valor de la venta" )
venta = float( venta_ )

# Procedimiento
sueldo = 600
sueldo_mensual = 0

if (venta >= 1 and venta <= 1000):
  comision = venta * 0.05
  sueldo_mensual = sueldo + comision
elif (venta > 1000 and venta <= 5000):
  comision = venta * 0.09
  sueldo_mensual = sueldo + comision
elif (venta > 5000):
  comision = venta * 0.12
  bono = sueldo * 0.10
  sueldo_mensual = sueldo + comision + bono

# Salida
print( "El sueldo a pagar es: " + str(sueldo_mensual) )
```
